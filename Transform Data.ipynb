# Insert all rows as per `new_df` from `df` using the above trick and loop

```python
crops_index = dict(zip(crops, list(zip(list(range(5, 72, 3)), list(range(8, 75, 3))))))
crops_index
```

```python
dist_codes = []
years = []
state_codes = []
state_names = []
dist_names = []
cropss = []
areas = []
productions = []
yeilds = []

for each_row in range(df.shape[0]):
    dist_code, year, state_code, state_name, dist_name = df.iloc[each_row, :5]
    for each_crop in crops:
        crop = each_crop
        area, production, yeild = df.iloc[each_row, crops_index[each_crop][0] : crops_index[each_crop][1]]
        
        dist_codes.append(dist_code)
        years.append(year)
        state_codes.append(state_code)
        state_names.append(state_name)
        dist_names.append(dist_name)
        cropss.append(crop)
        areas.append(area)
        productions.append(production)
        yeilds.append(yeild)
```

```python
new_df = pd.DataFrame({
            "Dist Code" : dist_codes,
            "Year" : years,
            "State Code" : state_codes,
            "State Name" : state_names,
            "Dist Name" : dist_names,
            "Crop" : cropss,
            "Area (1000ha.)" : areas,
            "Production (1000tons)" : productions,
            "Yield (Kg. per ha.)" : yeilds
            
        })
```
